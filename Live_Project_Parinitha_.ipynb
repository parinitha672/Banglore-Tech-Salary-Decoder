```python
#convert Current_CTC
def parse_ctc(ctc):
    if pd.isna(ctc):
        return np.nan

    ctc = str(ctc).replace("₹", "").replace("LPA", "").replace(",", "").strip()

    try:
        value = float(ctc)


        if value > 100000:
            value = value / 100000

        return value
    except:
        return np.nan
```

```python
df["current_CTC_clean"]=df["currenct_ctc"].apply(parse_ctc)
```

```python
print(df["current_CTC_clean"])==df["currenct_ctc"].apply(parse_ctc)
```

```python
# Convert Previous_CTC values into numeric LPA format
df["previous_CTC_Clean"] = df["previous_ctc"].apply(parse_ctc)
```

```python
# Remove ₹ symbol
df["currenct_ctc"] = df["currenct_ctc"].astype(str).str.replace("₹","", regex=False)

# Remove LPA text
df["currenct_ctc"] = df["currenct_ctc"].str.replace("LPA","", regex=False)

# Remove commas
df["currenct_ctc"] = df["currenct_ctc"].str.replace(",","", regex=False)

# Remove extra spaces
df["currenct_ctc"] = df["currenct_ctc"].str.strip()

# Convert to numeric
df["currenct_ctc"] = pd.to_numeric(df["currenct_ctc"], errors="coerce")
```

```python
# Clean the Current CTC column

df["currenct_ctc"] = df["currenct_ctc"].astype(str)

df["currenct_ctc"] = df["currenct_ctc"].str.replace("₹","",regex=False)
df["currenct_ctc"] = df["currenct_ctc"].str.replace("LPA","",regex=False)
df["currenct_ctc"] = df["currenct_ctc"].str.replace(",","",regex=False)
df["currenct_ctc"] = df["currenct_ctc"].str.strip()

df["currenct_ctc"] = pd.to_numeric(df["currenct_ctc"],errors="coerce")

# Convert salary in rupees to LPA
df.loc[df["currenct_ctc"]>1000,"currenct_ctc"]=df.loc[df["currenct_ctc"]>1000,"currenct_ctc"]/100000

print(df["currenct_ctc"].head())
```

```python
# Find median salary for each company type

company_salary=df.groupby("company_type")["currenct_ctc"].median()

print(company_salary)
```

```python
# Calculate median salary for each role

role_median=df.groupby('Role ')["currenct_ctc"].transform("median")

# Difference between employee salary and role median

df["salary_gap"]=role_median-df["currenct_ctc"]

# Top 10 underpaid employees

underpaid=df.sort_values(by="salary_gap",ascending=False).head(10)

print(underpaid[["employee_id",'Role ',"currenct_ctc","salary_gap"]])
```

```python
# Task 4 - Build the Final Printed Report

print("="*70)
print("           BANGALORE TECH SALARY DECODER")
print("      Built by Parinitha R | The Unlox Academy")
print("="*70)

print("\nDataset : Bangalore Tech Salary Dataset")
print("Period  : 2024 Employment Snapshot")

print("\n" + "-"*70)
print("MEDIAN CTC BY ROLE")
print("-"*70)

role_report = df.groupby('Role')["currenct_ctc"].median().sort_values(ascending=False)

for role, salary in role_report.items():
    print(f"{role:<25} {salary:.2f} LPA")


print("\n" + "-"*70)
print("SDE BACKEND CTC BY EXPERIENCE")
print("-"*70)

sde = df[df['Role'] == "SDE Backend"].copy()

bins=[0,1,3,5,100]
labels=["0-1","2-3","4-5","6+"]

sde.loc[:, "Experience_Group"] = pd.cut(
    sde["years_exp"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

exp_report=sde.groupby("Experience_Group", observed=False)["currenct_ctc"].median()

for exp, salary in exp_report.items():
    print(f"{exp:<10} {salary:.2f} LPA")


print("\n" + "-"*70)
print("COMPANY TYPE MEDIAN SALARY")
print("-"*70)

company_report=df.groupby("company_type")["currenct_ctc"].median()

for company,salary in company_report.items():
    print(f"{company:<20} {salary:.2f} LPA")


print("\n" + "-"*70)
print("TOP 5 UNDERPAID PROFESSIONALS")
print("-"*70)

top5=underpaid.head(5)

for index,row in top5.iterrows():
    print(f"{row['employee_id']:<10} {row['Role ']:<20} {row['currenct_ctc']:.2f} LPA")


print("="*70)
print("END OF REPORT")
print("="*70)
```
